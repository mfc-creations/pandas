```python
import pandas as pd
df = pd.read_csv("weather_data.csv",parse_dates=['day'])
type(df.day[0])
df
```

```python
df.set_index('day',inplace=True)
df
```

```python
# fillna
# Fill all NaN with one specific value
```

```python
new_df = df.fillna(0)
new_df
```

```python
# Fill na using column names and dict
new_df = df.fillna({
        'temperature': 0,
        'windspeed': 0,
        'event': 'No Event'
    })
new_df
```

```python
# Use method to determine how to fill na values
new_df = df.fillna(method="ffill")
new_df
```

```python
new_df = df.fillna(method="bfill")
new_df
```

```python
# Use of axis
new_df = df.fillna(method="bfill", axis="columns") # axis is either "index" or "columns"
new_df
```

```python
# limit parameter
new_df = df.fillna(method="ffill",limit=1)
new_df
```

```python
# interpolate
new_df = df.interpolate()
new_df
```

```python
new_df = df.interpolate(method="time") 
new_df
```

```python
# dropna
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how='all')
new_df
```

```python
new_df = df.dropna(thresh=1)
new_df
```

```python
# Inserting Missing Dates
dt = pd.date_range("01-01-2017","01-11-2017")
idx = pd.DatetimeIndex(dt)
df.reindex(idx)
```
